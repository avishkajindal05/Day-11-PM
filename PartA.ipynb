Student Management System (with Logging)

```python
import logging

logging.basicConfig(
    filename="student_errors.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

records = []

VALID_SUBJECTS = {"English", "Math", "Physics"}

def validate_student(name, subject, marks):

    if not name.strip():
        raise ValueError("Student name cannot be empty")

    if subject not in VALID_SUBJECTS:
        raise ValueError(f"Invalid subject: {subject}")

    if not (0 <= marks <= 100):
        raise ValueError("Marks must be between 0 and 100")


def add_student():

    try:
        name = input("Enter name: ").strip()
        subject = input("Enter subject (English, Math, Physics): ").strip()
        marks = int(input("Enter marks: "))

        validate_student(name, subject, marks)

        for r in records:
            if r[0] == name and r[1] == subject:
                raise ValueError("Duplicate student record")

    except ValueError as e:
        print(f"Input error: {e}")
        logging.error(f"Add student failed: {e}")

    except Exception as e:
        print("Unexpected system error occurred.")
        logging.error(f"Unexpected error: {e}")

    else:
        records.append([name, subject, marks])
        print("Student added successfully")

    finally:
        print("Operation finished.\n")


def save_records():

    try:
        with open("students.txt", "w") as f:
            for r in records:
                f.write(f"{r[0]}, {r[1]}, {r[2]}\n")

    except OSError as e:
        print("Could not save records.")
        logging.error(f"File write error: {e}")

    else:
        print("Records saved successfully.")

    finally:
        print("Save attempt completed.")


def menu():

    while True:

        try:
            choice = int(input("""
Student System
1 Add student
2 Save records
3 Exit

Enter choice: """))

            if choice not in [1,2,3]:
                raise ValueError("Choice must be between 1 and 3")

        except ValueError as e:
            print(f"Invalid input: {e}")
            continue

        else:
            if choice == 1:
                add_student()

            elif choice == 2:
                save_records()

            elif choice == 3:
                print("Exiting program")
                break

        finally:
            print("Menu cycle completed.")
```

Password Analyzer (Exception Safe)

```python
import random
import string

SPECIAL_CHARS = "!@#$%^&*"
MIN_LENGTH = 8
MAX_LENGTH = 32


def validate_length(length):

    if length < MIN_LENGTH or length > MAX_LENGTH:
        raise ValueError("Password length must be between 8 and 32")


def generate_password():

    while True:

        try:
            length = int(input("Enter password length (8-32): "))

            validate_length(length)

        except ValueError as e:
            print(f"Invalid input: {e}")

        else:
            char_pool = string.ascii_letters + string.digits + SPECIAL_CHARS
            password = "".join(random.choice(char_pool) for _ in range(length))

            print("Generated password:", password)
            return password

        finally:
            print("Password generation attempt finished.\n")
```

Product Catalog Search Tool

```python
catalog = {
    "SKU001": {"name": "Laptop", "price": 65000, "category": "electronics"},
    "SKU002": {"name": "Sneakers", "price": 4000, "category": "clothing"},
    "SKU003": {"name": "Python Book", "price": 900, "category": "books"},
}

VALID_CATEGORIES = {"electronics", "clothing", "books"}


def validate_price_range(min_price, max_price):

    if min_price < 0 or max_price < 0:
        raise ValueError("Prices cannot be negative")

    if min_price > max_price:
        raise ValueError("Min price cannot exceed max price")


def search_price_range():

    try:
        min_price = float(input("Enter minimum price: "))
        max_price = float(input("Enter maximum price: "))

        validate_price_range(min_price, max_price)

    except ValueError as e:
        print(f"Invalid input: {e}")

    else:
        results = {
            sku: product
            for sku, product in catalog.items()
            if min_price <= product["price"] <= max_price
        }

        print("Matching products:", results)

    finally:
        print("Search completed.\n")
```
